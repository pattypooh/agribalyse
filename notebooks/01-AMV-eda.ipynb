## Normalize variable names

```python
#Normalize column names to have them without accents and spaces
norm_cols=syn_df.columns.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').str.replace("-","").str.replace(":","").str.replace(" ","_").str.replace("'","")
norm_cols
```

```python
#Get rid of any details that are in brackets
norm_cols2=[colname.split('_(')[0] for colname in norm_cols]
norm_cols2
```

```python
#Rename columns
syn_df.columns=norm_cols
```
